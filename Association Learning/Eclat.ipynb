Market basket Optimization will be carried out in this project.

Problem Statement : Find association rules to find if customers buy "A" product then they are going to most probably buy "B" product and help the particular shop or market to optimize or maximize the sales using Eclat Model

As far as working of Eclat Model is concerned we will be making our conclusions based on the values of support where as we used Lift and Confidence values in Apriori

# Importing the dataset and libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataSet = pd.read_csv("Market_Basket_Optimisation.csv", header = None)
```

```python
dataSet
```

# Data Preprocessing

```python
# we are supposed to have a list of transactions
transactions = []

for i in range(dataSet.shape[0]):
  # transactions.append([j for j in dataSet.iloc[i,:].values if str(j)!= "nan"])
  transactions.append([str(dataSet.values[i,j]) for j in range(0,20)])
```

```python
transactions
```

# Training the Eclat model on dataset

```python
from apyori import apriori
```

```python
# imagine we are targetting in such a way that we have a deal : buy 1 get 1 free, hence min and max lengths are 2
# these lengths may vary based on the requirements
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

```python
rules
```

# Displaying the first results

```python
results = list(rules)
results
```

```python
results[2]
```

# Displaying Results in Pandas Data Frame

```python
def inspect(results):
  lhs = [tuple(result[2][0][0])[0] for result in results]
  rhs = [tuple(result[2][0][1])[0] for result in results]
  supports = [result[1] for result in results]
  return list(zip(lhs,rhs, supports))
resultsDataFrame = pd.DataFrame(inspect(results), columns =['Product 1', 'Product 2', 'Support'])
```

```python
inspect(results)
```
