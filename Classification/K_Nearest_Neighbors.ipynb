#Import libraries and dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataSet = pd.read_csv('Social_Network_Ads.csv')
```

```python
dataSet
```

```python
# Problem Statemen is to predict whether the asset is purchased or not
```

```python
x = dataSet.iloc[:,:-1].values
y = dataSet.iloc[:, -1].values
x
```

#Splitting the Dataset

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, random_state = 7)
```

#Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train) 
x_test = sc.transform(x_test)
print(x_train)
```

#Fitting the model on our data set

```python
from sklearn.neighbors import KNeighborsClassifier
KNNModel = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p=2)
```

```python
KNNModel.fit(x_train, y_train)
```

```python
KNNModel.score(x_train, y_train)
# hence we around 91% training accuracy
```

```python
KNNModel.score(x_test, y_test)
# hence we around 92% testing accuracy
```

#Prediciting new results

```python
age = int(input("Enter Age : "))
salary = int(input("Enter Salary : "))
prediction = KNNModel.predict(sc.transform([[age, salary]]))
if prediction[0] == 1:
  print("You can purchase")
else:
  print("You cannot purchase")
```

```python
# as per above examples we are able to experience that model is working properly
```

#Predicting on test data set

```python
y_pred = KNNModel.predict(x_test)
```

```python

y_pred
```

```python
y_test
```

```python
from sklearn.metrics import jaccard_score
# Jaccard score is the ratio of intersection of actual and predicted outputs to union of actual and predicted outputs
jaccard_score(y_test, y_pred,pos_label=0)
```

```python
# to compare the prediceted and actual outputs let us concatenate both of em and print so that it is easy to check ;)
np.set_printoptions(precision = 2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

#Confusion Matrix

```python
from sklearn.metrics import classification_report, confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
# we get accuracy of almost 93% 
```

```python
import seaborn as sns
```

```python
sns.heatmap(cm, annot=True)
```

#Visualisation of training dataset


```python
from matplotlib.colors import ListedColormap

markers = ('x', 'o', '^', 'v')
colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
cmap = ListedColormap(colors[:len(np.unique(y_train))])

    # plot the decision surface
x1_min, x1_max = x_train[:, 0].min() - 1,  x_train[:, 0].max() + 1    
x2_min, x2_max =  x_train[:, 1].min() - 1, x_train[:, 1].max() + 1
xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, 0.02),
np.arange(x2_min, x2_max, 0.02))
Z = KNNModel.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
Z = Z.reshape(xx1.shape)
plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
plt.xlim(xx1.min(), xx1.max())
plt.ylim(xx2.min(), xx2.max())

for idx, cl in enumerate(np.unique(y_train)):
    plt.scatter(x=x_train[y_train == cl, 0], y=x_train[y_train == cl, 1],
                alpha=0.4, c=cmap(idx),
                marker=markers[idx], label=cl)
```
