#Import libraries and dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataSet = pd.read_csv('Social_Network_Ads.csv')
```

```python
dataSet
```

```python
# Problem Statemen is to predict whether the asset is purchased or not
```

```python
x = dataSet.iloc[:,:-1].values
y = dataSet.iloc[:, -1].values
x
```

#Splitting the Dataset

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, random_state = 7)
```

#Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train) 
x_test = sc.transform(x_test)
print(x_train)
```

#Fitting the model on our data set

```python
from sklearn.linear_model import LogisticRegression
logisticModel = LogisticRegression()
```

```python
logisticModel.fit(x_train, y_train)
```

```python
logisticModel.score(x_train, y_train)
# hence we around 85% training accuracy
```

```python
logisticModel.score(x_test, y_test)
# hence we around 81% testing accuracy
```

#Predicting on test data set

```python
y_pred = logisticModel.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import jaccard_score
jaccard_score(y_test, y_pred,pos_label=0)
```

```python
# to compare the prediceted and actual outputs let us concatenate both of em and print so that it is easy to check ;)
np.set_printoptions(precision = 2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

#Making Confusion matrix

```python
from sklearn.metrics import classification_report, confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

#Visualisation of training dataset


```python
from matplotlib.colors import ListedColormap
x_set, y_set = sc.inverse_transform(x_train), y_train
X1, X2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 10, stop = x_set[:, 0].max() + 10, step = 0.25),
np.arange(start = x_set[:, 1].min() - 1000, stop = x_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, logisticModel.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),alpha = 0.5, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
  plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
